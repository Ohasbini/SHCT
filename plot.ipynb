```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rc('font', size=12)          # controls default text sizes
plt.rc('axes', titlesize=12)     # fontsize of the axes title
plt.rc('axes', labelsize=12)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=12)    # fontsize of the tick labels
plt.rc('ytick', labelsize=12)    # fontsize of the tick labels
plt.rc('legend', fontsize=12)    # legend fontsize
plt.rc('figure', titlesize=12)  # fontsize of the figure title
```

```python
# Define the file path
file_path = "server heating power"

# Initialize a list to collect the data
data = []

# Open the file and read it line by line
with open(file_path, 'r') as file:
    for line in file:
        # Replace commas with periods
        line = line.replace(",", "")
        # Split the line into values and convert them to float
        values = [float(value) for value in line.split()]
        # Extend the data list with the values
        data.extend(values)
        
# Convert the data list to a NumPy array
server_heating_power = np.array(data)
Time = np.linspace(0,24,len(server_heating_power))

print("Maximum Power:", np.max(server_heating_power), "kW")
```

```python
#plt.figure(figsize=(10, 6))
plt.figure(figsize=(10, 4.5))
plt.grid()
plt.plot(Time, server_heating_power, label='Server Heating Power')
plt.xticks(np.linspace(0,24,13))
#plt.xlim()
#plt.ylim(1, 9.5)
#plt.yticks(np.linspace(0,90,10))
plt.xlabel('Time [h]')
plt.ylabel('[kW]')
plt.title("Server Heating Power")
plt.legend()

#plt.savefig("P_Pilot_req_plot.pdf")
plt.show()
```

## Ambient Temperature Plots

```python
plt.figure(figsize=(10, 4.5))
plt.grid(alpha=0.5)

# Define the file path
file_path = "ambient temperature spring"
# Initialize a list to collect the data
data = []
# Open the file and read it line by line
with open(file_path, 'r') as file:
    for line in file:
        # Replace commas with periods
        line = line.replace(",", "")
        # Split the line into values and convert them to float
        values = [float(value) for value in line.split()]
        # Extend the data list with the values
        data.extend(values)
        
# Convert the data list to a NumPy array
ambient_temp = np.array(data)
Time = np.linspace(0,24,len(ambient_temp))

#print("Maximum Temperature:", np.max(ambient_temp), "C")

# Find the index where the maximum occurs
max_temp = np.max(ambient_temp)
max_index = np.argmax(ambient_temp)
max_time = Time[max_index]

# Marker at the maximum point
plt.plot(max_time, max_temp, 'ro', markersize=6)   # red circle

# annotation with arrow
plt.annotate(f'{max_temp:.1f} °C',
             xy=(max_time, max_temp),
             xytext=(max_time + 0, max_temp + 4),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=5))


plt.plot(Time, ambient_temp, color="tab:green", label='$T_{amb,Spring}$')


# Define the file path
file_path = "ambient temperature summer"
# Initialize a list to collect the data
data = []
# Open the file and read it line by line
with open(file_path, 'r') as file:
    for line in file:
        # Replace commas with periods
        line = line.replace(",", "")
        # Split the line into values and convert them to float
        values = [float(value) for value in line.split()]
        # Extend the data list with the values
        data.extend(values)
        
# Convert the data list to a NumPy array
ambient_temp = np.array(data)

# Find the index where the maximum occurs
max_temp = np.max(ambient_temp)
max_index = np.argmax(ambient_temp)
max_time = Time[max_index]

# Marker at the maximum point
plt.plot(max_time, max_temp, 'ro', markersize=6)   # red circle

# annotation with arrow
plt.annotate(f'{max_temp:.1f} °C',
             xy=(max_time, max_temp),
             xytext=(max_time + 0.5, max_temp + 0.5),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=5))
plt.plot(Time, ambient_temp, color="tab:red", label='$T_{amb,Summer}$')


# Define the file path
file_path = "ambient temperature fall"
# Initialize a list to collect the data
data = []
# Open the file and read it line by line
with open(file_path, 'r') as file:
    for line in file:
        # Replace commas with periods
        line = line.replace(",", "")
        # Split the line into values and convert them to float
        values = [float(value) for value in line.split()]
        # Extend the data list with the values
        data.extend(values)
        
# Convert the data list to a NumPy array
ambient_temp = np.array(data)

# Find the index where the maximum occurs
max_temp = np.max(ambient_temp)
max_index = np.argmax(ambient_temp)
max_time = Time[max_index]

# Marker at the maximum point
plt.plot(max_time, max_temp, 'ro', markersize=6)   # red circle

# annotation with arrow
plt.annotate(f'{max_temp:.1f} °C',
             xy=(max_time, max_temp),
             xytext=(max_time + 0, max_temp - 4),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=5))
plt.plot(Time, ambient_temp, color="tab:orange", label='$T_{amb,Fall}$')


# Define the file path
file_path = "ambient temperature winter"
# Initialize a list to collect the data
data = []
# Open the file and read it line by line
with open(file_path, 'r') as file:
    for line in file:
        # Replace commas with periods
        line = line.replace(",", "")
        # Split the line into values and convert them to float
        values = [float(value) for value in line.split()]
        # Extend the data list with the values
        data.extend(values)
        
# Convert the data list to a NumPy array
ambient_temp = np.array(data)

# Find the index where the maximum occurs
max_temp = np.max(ambient_temp)
max_index = np.argmax(ambient_temp)
max_time = Time[max_index]

# Marker at the maximum point
plt.plot(max_time, max_temp, 'ro', markersize=6)   # red circle

# annotation with arrow
plt.annotate(f'{max_temp:.1f} °C',
             xy=(max_time, max_temp),
             xytext=(max_time + 0.5, max_temp + 0.5),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=5))
plt.plot(Time, ambient_temp, color="tab:blue", label='$T_{amb,Winter}$')








plt.xticks(np.linspace(0,24,13))
#plt.xlim()
plt.ylim(0, 38)
#plt.yticks(np.linspace(0,90,10))
plt.xlabel('Time [h]')
plt.ylabel('Temperature [°C]')
plt.title("Ambient Temperature in Different Seasons")
plt.legend()

#plt.savefig("P_Pilot_req_plot.pdf")
plt.show()
```
