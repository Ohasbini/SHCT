```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rc('font', size=12)          # controls default text sizes
plt.rc('axes', titlesize=12)     # fontsize of the axes title
plt.rc('axes', labelsize=12)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=12)    # fontsize of the tick labels
plt.rc('ytick', labelsize=12)    # fontsize of the tick labels
plt.rc('legend', fontsize=12)    # legend fontsize
plt.rc('figure', titlesize=12)  # fontsize of the figure title
```

```python
# Define the file path
file_path = "server heating power"

# Initialize a list to collect the data
data = []

# Open the file and read it line by line
with open(file_path, 'r') as file:
    for line in file:
        # Replace commas with periods
        line = line.replace(",", "")
        # Split the line into values and convert them to float
        values = [float(value) for value in line.split()]
        # Extend the data list with the values
        data.extend(values)
        
# Convert the data list to a NumPy array
server_heating_power = np.array(data)
Time = np.linspace(0,24,len(server_heating_power))
```

```python
#plt.figure(figsize=(10, 6))
plt.figure(figsize=(10, 4.5))
plt.grid()
plt.plot(Time, server_heating_power, label='Server Heating Power')
plt.xticks(np.linspace(0,24,13))
#plt.xlim()
#plt.ylim(1, 9.5)
#plt.yticks(np.linspace(0,90,10))
plt.xlabel('Time [s]')
plt.ylabel('[kW]')
plt.title("Server Heating Power")
plt.legend()

#plt.savefig("P_Pilot_req_plot.pdf")
plt.show()
```
